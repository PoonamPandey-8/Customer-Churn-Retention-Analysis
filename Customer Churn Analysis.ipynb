```python
df.columns = df.columns.str.strip().str.lower()
```

```python
df = df.rename(columns={
    "customerid": "customer_id",
    "seniorcitizen": "senior_citizen",
    "phoneservice": "phone_service",
    "multiplelines": "multiple_lines",
    "internetservice": "internet_service",
    "onlinesecurity": "online_security",
    "onlinebackup": "online_backup",
    "deviceprotection": "device_protection",
    "techsupport": "tech_support",
    "streamingtv": "streaming_tv",
    "streamingmovies": "streaming_movies",
    "paperlessbilling": "paperless_billing",
    "paymentmethod": "payment_method",
    "monthlycharges": "monthly_charges",
    "totalcharges": "total_charges",
    "Churn": "churn"
})
```

```python
df.columns
```

```python
df.dtypes
```

```python
df["senior_citizen"] = df["senior_citizen"].map({
    0: "No",
    1: "Yes"
})
```

```python
df["senior_citizen"].value_counts()
```

```python
df["total_charges"].unique()
```

```python
df["total_charges"] = df["total_charges"].replace(" ", np.nan)
df['total_charges'] = pd.to_numeric(df['total_charges'], errors='coerce')
```

```python
df["total_charges"].dtype
```

```python
df.isnull().sum()
```

```python
df[df["total_charges"].isnull()]
```

```python
df[df["total_charges"].isnull()][
    ["customer_id", "tenure", "monthly_charges", "total_charges", "churn"]
]
```

```python
df.loc[
    df["total_charges"].isnull() & (df["tenure"] == 0),
    "total_charges"
] = 0
```

```python
cat_cols = df.select_dtypes(include="object").columns
cat_cols
```

```python
for col in cat_cols:
    print(col)
    print(df[col].unique())
    print()
```

```python
df["tenure_group"] = pd.cut(
    df["tenure"],
    bins=[0, 12, 24, 48, 60, 72],
    labels=["0-12 months", "13-24 months", "25-48 months", "49-60 months", "61-72 months"],
    include_lowest=True
)
```

```python
tenure_order = {
    "0-12 months": 1,
    "13-24 months": 2,
    "25-48 months": 3,
    "49-60 months": 4,
    "61-72 months": 5
}

df["tenure_group_order"] = df["tenure_group"].astype(str).map(tenure_order)
```

```python
df["monthly_charge_group"] = pd.cut(
    df["monthly_charges"],
    bins=[0, 35, 70, 100, 120],
    labels=["Low", "Medium", "High", "Very High"]
)
```

```python
charge_order = {
    "Low": 1,
    "Medium": 2,
    "High": 3,
    "Very High": 4
}

df["monthly_charge_group_order"] = df["monthly_charge_group"].astype(str).map(charge_order)
```

```python
df[["tenure_group_order", "monthly_charge_group_order"]].isnull().sum()
```

```python
df[["tenure_group_order", "monthly_charge_group_order"]].head()
```

```python
df["tenure_group_order"] = df["tenure_group_order"].astype("Int64")
df["monthly_charge_group_order"] = df["monthly_charge_group_order"].astype("Int64")
```

```python
df["churn_flag"] = df["churn"].map({
    "Yes": 1,
    "No": 0
})
```

```python
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
categorical_cols = df.select_dtypes(include="object").columns

print("Numerical columns:")
print(numeric_cols)

print("Categorical columns:")
print(categorical_cols)
```
